**9. Write a Python program to create a heatmap to visualize the correlation between different subjects in a student's report card?**
___

```python
import matplotlib.pyplot as plt
import numpy as np
# Data
subjects = ['Math', 'Science', 'English', 'History']
correlation_matrix = np.array([[1.0, 0.8, 0.6, 0.4],
                                 [0.8, 1.0, 0.7, 0.5],
                                 [0.6, 0.7, 1.0, 0.9],
                                 [0.4, 0.5, 0.9, 1.0]])

fig, ax = plt.subplots()
im = ax.imshow(correlation_matrix, cmap='hot')

ax.set_xticks(np.arange(len(subjects)))
ax.set_yticks(np.arange(len(subjects)))
ax.set_xticklabels(subjects)
ax.set_yticklabels(subjects)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(subjects)):
    for j in range(len(subjects)):
        text = ax.text(j, i, correlation_matrix[i, j], ha="center", va="center", color="w")

ax.set_title("Correlation Between Subjects")
fig.tight_layout()
plt.show()
```

**11. Write a Python program to create a figure with multiple subplots, each displaying a different type of plot?**
___

```python
import matplotlib.pyplot as plt
import numpy as np
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axs = plt.subplots(3)
fig.suptitle('Multiple Subplots')

axs[0].plot(x, y1)
axs[0].set_title('Sine')

axs[1].plot(x, y2)
axs[1].set_title('Cosine')

axs[2].plot(x, y3)
axs[2].set_title('Tangent')

plt.tight_layout()
plt.show()
```

**17. Using the summer.csv dataset create a stacked bar chart to visualize the number of Gold, Silver, and Bronze medals won by the top 5 countries.**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('summer.csv')
top_countries = df['Country'].value_counts().head(5).index
top_countries_medals = df[df['Country'].isin(top_countries)]
medal_counts = top_countries_medals.groupby(['Country', 'Medal']).size().unstack(fill_value=0)
medal_counts = medal_counts[['Gold', 'Silver', 'Bronze']]

medal_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Medal Counts of Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.show()
```

**18. Using the summer.csv dataset create an area plot to visualize the trend of the number of athletes from different countries over the years?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('summer.csv')
top_countries = df['Country'].value_counts().head(5).index
athlete_counts = df[df['Country'].isin(top_countries)].groupby(['Year', 'Country'])['Athlete'].nunique().unstack()

athlete_counts.plot(kind='area', stacked=False, figsize=(12, 6))
plt.title('Number of Athletes from Top 5 Countries Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.show()
```

**19. Using the Titanic dataset create a bar chart to visualize the number of passengers in each class (1st, 2nd, 3rd).**

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_df = pd.read_csv('titanic.csv')
class_counts = titanic_df['pclass'].value_counts().sort_index()

plt.bar(class_counts.index, class_counts.values)
plt.title('Number of Passengers in Each Class on the Titanic')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks([1, 2, 3], ['1st', '2nd', '3rd'])
plt.show()
```

**20. Using the Titanic dataset create a figure with two subplots: the first subplot should display a line plot of the 'age' column, and the second subplot should display a bar chart of the 'sex' column showing the count of each gender.**
___

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_df = pd.read_csv('titanic.csv')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(titanic_df['age'].dropna())
ax1.set_title('Age of Titanic Passengers')
ax1.set_xlabel('Passenger Index')
ax1.set_ylabel('Age')

sex_counts = titanic_df['sex'].value_counts()
ax2.bar(sex_counts.index, sex_counts.values)
ax2.set_title('Gender Distribution on the Titanic')
ax2.set_xlabel('Sex')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```
